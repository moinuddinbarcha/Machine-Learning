```python

import pandas as pd
import numpy as np
import sklearn

df = pd.read_csv('/content/drive/MyDrive/Boston.csv')
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# Assuming df is already loaded with the correct data as in the previous code

x = np.array(df.iloc[:, 0:13])
y = np.array(df.loc[:, 'medv'])  # Corrected the indexing here using loc

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor



# load the KNN model
knn= KNeighborsRegressor(3)
knn.fit(x_train, y_train)

pred_knn=knn.predict(x_test)
```

**Hyperparamater Tuning**

```python
import numpy as np
import sklearn
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Assuming x_train, y_train, x_test, y_test are already defined
# Initialize an empty list to store MSE values
mse_values = []

for i in range(1, 50):
    model = KNeighborsRegressor(n_neighbors=i)
    model.fit(x_train, y_train)

    pred_model = model.predict(x_test)

    mse = mean_squared_error(y_test, pred_model)
    mse_values.append(mse)

    print(f'MSE for k={i}: {mse}')

# Find the index of the minimum MSE
best_k = np.argmin(mse_values) + 1  # +1 because indexing starts from 0
best_mse = mse_values[best_k - 1]  # -1 to account for 0-based indexing

print(f'Best k value: {best_k}')
print(f'Lowest MSE: {best_mse}')
```

```python
import sklearn.metrics

knn_mse = sklearn.metrics.mean_squared_error(y_test, pred_knn, squared=False)
print('error for linear regression:', knn_mse)
```

```python
x = new.iloc[:, [0, 1]].values
y = new.iloc[:, 2].values

```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25, random_state=5)
```

```python
# scaling using standered scaler for normal distribution

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

**BNuilding the model using RBF kernal**

```python
from sklearn.svm import SVC
classifier_rbf = SVC(kernel = 'rbf', random_state=0)
classifier_rbf.fit(x_train, y_train)

y_rbf_pred = classifier_rbf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_rbf = classification_report(y_test, y_rbf_pred)
print(cr_rbf)
```
